**DATA PREPROCESING**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv('loan_data_2007_2014.csv')
# --- 1. Drop kolom yang tidak berguna atau terlalu banyak missing ---
drop_cols = [
    'Unnamed: 0', 'id', 'member_id', 'url', 'desc', 'title',
    'emp_title', 'zip_code', 'next_pymnt_d',  # redundant or tidak informatif
    # Kolom dengan >90% missing
    'annual_inc_joint', 'dti_joint', 'verification_status_joint',
    'mths_since_last_major_derog', 'tot_coll_amt', 'tot_cur_bal',
    'open_acc_6m', 'open_il_6m', 'open_il_12m', 'open_il_24m',
    'mths_since_rcnt_il', 'total_bal_il', 'il_util',
    'open_rv_12m', 'open_rv_24m', 'max_bal_bc', 'all_util',
    'inq_fi', 'total_cu_tl', 'inq_last_12m'
]
df = df.drop(columns=drop_cols)

# --- 2. Simplifikasi target (loan_status) menjadi biner ---
# 1 = berisiko (Charged Off / Default), 0 = aman (Fully Paid)
df = df[df['loan_status'].isin(['Fully Paid', 'Charged Off', 'Default'])].copy()
df['loan_status'] = df['loan_status'].apply(lambda x: 1 if x in ['Charged Off', 'Default'] else 0)

# --- 3. Imputasi missing values (numerik dengan median, kategorikal dengan modus) ---
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# --- 4. Format ulang kolom 'term' jadi angka (36 bulan -> 36) ---
df['term'] = df['term'].str.extract('(\d+)').astype(int)

# --- 5. Format ulang 'emp_length' jadi angka (10+ years → 10, < 1 year → 0.5, n/a → 0) ---
def parse_emp_length(x):
    if pd.isnull(x) or x == 'n/a':
        return 0
    if '<' in x:
        return 0.5
    if '+' in x:
        return 10
    return float(x.split()[0])
df['emp_length'] = df['emp_length'].apply(parse_emp_length)

# --- 6. Ubah kolom waktu menjadi datetime (jika diperlukan nanti untuk EDA) ---
df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%y')  # fix format: Dec-11 → %b-%y
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], format='%b-%y')
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], format='%b-%y')
df['last_credit_pull_d'] = pd.to_datetime(df['last_credit_pull_d'], format='%b-%y')

# --- 7. Encoding variabel kategorikal ---
df = pd.get_dummies(df, columns=[
    'grade', 'sub_grade', 'home_ownership', 'verification_status',
    'purpose', 'application_type', 'initial_list_status', 'pymnt_plan'
], drop_first=True)

# --- 8. Feature engineering: usia kredit dalam bulan ---
df['credit_age_months'] = (
    (df['issue_d'].dt.year - df['earliest_cr_line'].dt.year) * 12 +
    (df['issue_d'].dt.month - df['earliest_cr_line'].dt.month)
).astype('Int64')

# --- 9. Split fitur dan target ---
X = df.drop(columns=['loan_status', 'issue_d', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d', 'addr_state'])
y = df['loan_status']

# --- 10. Split data untuk training dan testing ---
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

**MODELLING AND EVALUATION**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42)
}
```

```python
def evaluate_model(name, model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1] if hasattr(model, 'predict_proba') else None

    print(f"\n{name}")
    print("-"*len(name))
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test, y_proba) if y_proba is not None else "N/A")
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
for name, model in models.items():
    evaluate_model(name, model, X_train_scaled, X_test_scaled, y_train, y_test)
```
